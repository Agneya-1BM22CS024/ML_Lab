```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
data = yf.download(tickers, start="2024-01-01", end="2024-12-30",group_by='ticker')
print(data.head())
```

```python
data.shape
print(data.columns)
```

```python
hdfc_data = data['HDFCBANK.NS']
print(hdfc_data.head())
```

```python
hdfc_data['Daily Return'] = hdfc_data['Close'].pct_change()
hdfc_data.head()
```

```python
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
hdfc_data['Close'].plot(title='HDFC Stock Price')
plt.subplot(2,1,2)
hdfc_data['Daily Return'].plot(title='HDFC Daily Return')
plt.tight_layout()
plt.show()
```

```python
kotak_data = data['KOTAKBANK.NS']
kotak_data['Daily Return'] = kotak_data['Close'].pct_change()
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
kotak_data['Close'].plot(title='Kotak Closing Price')
plt.subplot(2,1,2)
kotak_data['Daily Return'].plot(title="Kotak Daily Return")
plt.tight_layout()
plt.show()
```

```python
icici_data = data['ICICIBANK.NS']
icici_data['Daily Return'] = icici_data['Close'].pct_change()
plt.figure(figsize=(12,6))
plt.subplot(2,1,1)
icici_data['Close'].plot(title='ICICI Closing Price')
plt.subplot(2,1,2)
icici_data['Daily Return'].plot(title="ICICI Daily Return")
plt.tight_layout()
plt.show()
```
